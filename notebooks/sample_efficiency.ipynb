```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Given data
data = {
    50: [70, 65, 80],
    100: [100, 100, 94],
    500: [100, 100, 100],
    10000: [100, 100, 100]
}
```

```python
# Convert the dictionary to a dataframe with long format
df = pd.DataFrame(data).melt(var_name='X', value_name='Y')

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='X', y='Y', errorbar='sd', marker='o', linewidth=2.5)

plt.xlabel('Number of training samples')
plt.xscale('log')
plt.ylabel('OOD accuracy')
#plt.title('Plot with Mean and Standard Deviation Highlighted Range')
plt.show()
```
