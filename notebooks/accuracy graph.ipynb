```python
sns.set_context("paper", font_scale=2)
# Reset the index to "melt" the dataframe for seaborn
df_melted = df.reset_index()

# Melt the DataFrame to have a single column for the model types and a column for the values
df_melted = df_melted.melt(id_vars=['Data set', 'Test type'], var_name='Model', value_name='Value')

df_melted['Value'] += .015

# Filter for 'test OOD structural' and plot
df_structural = df_melted[df_melted['Test type'] == 'test OOD structural']#[df_melted['Data set'] != 'car_cdr_seq']
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Model', y='Value', hue='Data set', data=df_structural)
plt.xticks(rotation=15)
plt.ylabel('Accuracy', weight='bold')
plt.xlabel('Model', weight='bold')
plt.title('Structural Out-of-Distribution\nGeneralization (Novel Nodes)', loc='left', fontsize=22)

save_legend_separately = False
ax.legend_.remove()  # remove legend
if save_legend_separately:
    # Get the current legend
    legend = ax.get_legend()
    # Create a new figure for the legend
    new_fig = plt.figure()
    new_ax = new_fig.add_subplot()
    new_ax.legend(legend.legendHandles, [x.get_text() for x in legend.get_texts()], title=legend.get_title().get_text())
    new_ax.axis('off')
    new_fig.show()

#sns.move_legend(ax, "upper right", bbox_to_anchor=(1, 1.4))
plt.tight_layout()
plt.savefig("test_OOD_structural.svg", format='svg')
plt.show()

# Filter for 'test OOD lexical' and plot
df_lexical = df_melted[df_melted['Test type'] == 'test OOD lexical']#[df_melted['Data set'] != 'car_cdr_seq']
plt.figure(figsize=(8,5))
ax = sns.barplot(x='Model', y='Value', hue='Data set', data=df_lexical)
plt.xticks(rotation=15)
plt.ylabel('Accuracy', weight='bold')
plt.xlabel('Model', weight='bold')
ax.legend_.remove()  # remove legend
plt.title('Lexical Out-of-Distribution Generalization (Novel Words)', fontsize=22)
plt.tight_layout()
plt.savefig("test_OOD_lexical.svg", format='svg')
plt.show()

```
