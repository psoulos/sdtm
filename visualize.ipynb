```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## PG vs TPR

```python
df = pd.read_csv('result.tsv', sep='\t', names=['op_ge', 'query_ge', 'gs_temp', 'seed', 'gamma', 'train_acc', 'val_acc'])

df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df['Method'] = 'TPR'
df.loc[df['gamma']==-1, 'Method'] = 'REINFORCE'
df.loc[df['gamma']>0, 'Method'] = 'REINFORCE+TPR (naive interpolation)'
df.loc[df['op_ge']=='sample_st', 'op_ge'] = 'Sample'
df.loc[df['op_ge']=='argmax_st', 'op_ge'] = 'Argmax'
df.loc[df['op_ge']=='gsst', 'op_ge'] = 'Gumbel'
df.loc[df['query_ge']=='sample_st', 'query_ge'] = 'Sample'
df.loc[df['query_ge']=='argmax_st', 'query_ge'] = 'Argmax'
df.loc[df['query_ge']=='gsst', 'query_ge'] = 'Gumbel'
df['op_query'] =  df['op_ge'] + '/' + df['query_ge']
```

```python
plt.figure(figsize=(17, 4))
ax = sns.barplot(data=df[df.gamma.isin([0, -1, 0.01])], hue='Method', y='val_acc', x='op_query', estimator=np.mean)
ax.set_xlabel('Sampling Methods (Operation, Query)')
ax.set_ylabel('Valid. Acc.')
ax.set_title('CAR_CDR_RCONS (fully discrete)')
```

## TPR PG vs HVI vs TPR

```python
## TPR loss
df = pd.read_json('pg_vs_hvi_vs_tpr_200epochs.json', lines=True)

df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df['Method'] = 'TPR'
df.loc[(df['gamma']==-1)&(df['aux_loss_type']=='pg'), 'Method'] = 'PG'
df.loc[(df['gamma']==-1)&(df['aux_loss_type']=='hvi'), 'Method'] = 'HVI'
df.loc[(df['gamma']==0.01)&(df['aux_loss_type']=='pg'), 'Method'] = 'PG+TPR (0.01)'
df.loc[(df['gamma']==0.01)&(df['aux_loss_type']=='hvi'), 'Method'] = 'HVI+TPR (0.01)'
df.loc[(df['gamma']==0.1)&(df['aux_loss_type']=='pg'), 'Method'] = 'PG+TPR (0.1)'
df.loc[(df['gamma']==0.1)&(df['aux_loss_type']=='hvi'), 'Method'] = 'HVI+TPR (0.1)'
df.loc[(df['gamma']==1)&(df['aux_loss_type']=='pg'), 'Method'] = 'PG+TPR (1.)'
df.loc[(df['gamma']==1)&(df['aux_loss_type']=='hvi'), 'Method'] = 'HVI+TPR (1.)'
df.loc[df['op_grad_estimator']=='sample_st', 'op_grad_estimator'] = 'Sample'
df.loc[df['op_grad_estimator']=='argmax_st', 'op_grad_estimator'] = 'Argmax'
df.loc[df['op_grad_estimator']=='gsst', 'op_grad_estimator'] = 'Gumbel'
df.loc[df['query_grad_estimator']=='sample_st', 'query_grad_estimator'] = 'Sample'
df.loc[df['query_grad_estimator']=='argmax_st', 'query_grad_estimator'] = 'Argmax'
df.loc[df['query_grad_estimator']=='gsst', 'query_grad_estimator'] = 'Gumbel'
df['op_query'] =  df['op_grad_estimator'] + '/' + df['query_grad_estimator']
df['opt_lr'] =  df['optimizer'] + '/' + df['lr'].astype(str)
```

```python
plt.figure(figsize=(18, 4.2))
ax = sns.barplot(data=df[df.gamma.isin([0, -1, 0.01, 0.1])], hue='Method', y='valid_acc', x='op_query', estimator=np.mean)
ax.set_xlabel('Sampling Methods (Operation, Query)')
ax.set_ylabel('Valid. Acc.')
ax.set_title('CAR_CDR_RCONS (fully discrete)')
#plt.legend([],[], frameon=False)
```
